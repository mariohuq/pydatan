```python
titanic['PassengerId'] = pd.to_numeric(titanic['PassengerId'], downcast='unsigned') # Convert argument to a numeric type.
titanic['Survived'] = pd.to_numeric(titanic['Survived'], downcast='unsigned')
titanic['Pclass'] = pd.to_numeric(titanic['Pclass'], downcast='unsigned')
titanic['Age'] = pd.to_numeric(titanic['Age'], downcast='float')
titanic['Sex'] = titanic['Sex'].astype('category') # Cast a pandas object to a specified dtype
titanic.info()
```
